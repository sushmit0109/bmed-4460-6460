```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
```

```python
#fName = "tissue.png"
#fName = "micro.png"
fName = "flower.jpg"

I = imageio.imread(fName)
if len(I.shape) > 2:
    I = I[:,:,0]
print(I.dtype)
print(I.shape)

print(I.min())
print(I.mean())
print(I.std())

plt.imshow(I, cmap='gray', vmin=0, vmax=255)
```

```python
hist, bins = np.histogram(I, bins=256, range=[0,256])
#
# 1. Write your own code to compute the histogram of an image 
# to replace the above line of statement
#
# Initialize
#
# hist = np.zeros(256,)
# ??????
#
# bins = np.arange(257)
#

cum_hist = np.cumsum(hist)

height, width = I.shape
norm_cum_hist = cum_hist / (height * width)
norm_hist = hist / hist.max()

#width = (bins[1] - bins[0])
center = (bins[:-1] + bins[1:]) / 2
plt.bar(center, norm_hist, align='center')
plt.plot(norm_cum_hist, color='r')
plt.show()
```

```python
hists_cdf = (norm_cum_hist * 255).astype("uint8")

# mapping
img_eq = hists_cdf[I]
plt.imshow(img_eq, cmap='gray')
```
